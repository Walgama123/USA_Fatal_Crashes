```python
# Dependencies
import json
import requests
import pandas as pd
```

```python
usa_states = pd.read_csv("../Resources/USAStates.csv")
usa_states
```

```python
list_url=[]
#Base url for the API
baseurl = 'https://crashviewer.nhtsa.dot.gov/CrashAPI/crashes/GetCaseList?'

#specify the number of vehicales 
minNumOfVehicles=1
maxNumOfVehicles=10
format='&format=json'
url_numvehicales=(f'&minNumOfVehicles={minNumOfVehicles}&maxNumOfVehicles={maxNumOfVehicles}')

# Create a DataFrame with a column named 'year' and the specified years
years = pd.DataFrame(data=[2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022], columns=['year'])

for row in usa_states.itertuples():
    url_states=(f'states={row.StateID}')
    for row in years.itertuples():
        url_yearpart=(f'&fromYear={row.year}&toYear={row.year}')
        final_url=(f'{baseurl}{url_states}{url_yearpart}{url_numvehicales}{format}')
        list_url.append(final_url)
    
list_url
```

```python
#convert url list to a dataframe to iterate
list_url_df=pd.DataFrame(data=list_url,columns=['url'])
#list_url_df
```

```python
#query = url + character
crash_list_full=[]
for row in list_url_df.itertuples():
    #print(f'getting records from {row.url}')
    response = requests.get(row.url)
    response_json = response.json()
    crash_list=response_json['Results'][0]
    crash_list_full=crash_list_full+crash_list
    #crash_list_df=pd.DataFrame(crash_list)
    #crash_list_df

#crash_list_full    
    
    
```

```python
crash_list_full_df=pd.DataFrame(crash_list_full)
crash_list_full_df
```

```python
#Clean the and arrange the data set 
#split tht CountyName name to county and county number
crash_list_full_df[['County', 'CountyNumber']] = crash_list_full_df['CountyName'].str.extract(r'([A-Z\s]+)\s\((\d+)\)')
crash_list_full_df.head(5)
```

```python
# Extract the timestamp from the 'DateTime' column
crash_list_full_df['CrashDateTime'] = crash_list_full_df['CrashDate'].str.extract(r'/Date\((\d+)-\d+\)/')

# Convert the timestamp to datetime format
crash_list_full_df['CrashDateTime'] = pd.to_datetime(crash_list_full_df['CrashDateTime'], unit='ms')
crash_list_full_df.head(5)
```

```python
#drop unwanted columns 
crash_list_full_df=crash_list_full_df.drop(['CountyName','CrashDate','State'],axis=1)
crash_list_full_df
```

```python
crash_list_full_df=crash_list_full_df[['St_Case','StateName','County','CountyNumber','CrashDateTime','Fatals','Peds','Persons','TotalVehicles']]
#crash_list_full_df
```

```python
crash_list_full_df = crash_list_full_df[crash_list_full_df['CrashDateTime'].dt.year != 2023]
```
